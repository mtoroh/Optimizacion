```python
# Optimización: Maximizar ganancia en producción de fertilizantes
# Variables: x1 = toneladas de F1, x2 = toneladas de F2
# Función objetivo: Z = 180x1 + 220x2

from scipy.optimize import linprog

# Coeficientes de la función objetivo (negados para maximizar)
costos = [-180, -220]

# Matriz de restricciones (lado izquierdo)
A = [
    [3, 4],  # Compuesto A
    [2, 1],  # Compuesto B
    [5, 3],  # Compuesto C
]

# Límites disponibles de cada compuesto
b = [120, 80, 150]

# No negatividad
bounds = [(0, None), (0, None)]

# Resolver con linprog (método HiGHS)
sol = linprog(costos, A_ub=A, b_ub=b, bounds=bounds, method='highs')

x1, x2 = sol.x
ganancia = -sol.fun

print("=" * 45)
print("  Fertilizantes - Solución óptima")
print("=" * 45)
print(f"  F1 producido: {x1:.2f} ton")
print(f"  F2 producido: {x2:.2f} ton")
print(f"  Ganancia máxima: ${ganancia:.2f}")
print()

# Verificar que se respetan los límites de insumos
usos = {"A": 3*x1 + 4*x2, "B": 2*x1 + 1*x2, "C": 5*x1 + 3*x2}
limites_disp = {"A": 120, "B": 80, "C": 150}

print("  Uso de compuestos:")
for comp, uso in usos.items():
    print(f"    Compuesto {comp}: {uso:.2f} / {limites_disp[comp]}")
print("=" * 45)
```

```python
# Dinámica discreta de producción: sillas y mesas semana a semana
# S(n+1) = 0.6*S(n) + 0.2*M(n) + 40
# M(n+1) = 0.1*S(n) + 0.5*M(n) + 20

sillas = float(input("Ingrese S_0 (sillas iniciales): "))
mesas  = float(input("Ingrese M_0 (mesas iniciales): "))
n_semanas = int(input("¿Cuántas semanas desea simular? "))

historial = [(0, sillas, mesas)]

for semana in range(1, n_semanas + 1):
    nuevas_sillas = 0.6 * sillas + 0.2 * mesas + 40
    nuevas_mesas  = 0.1 * sillas + 0.5 * mesas + 20
    sillas, mesas = nuevas_sillas, nuevas_mesas
    historial.append((semana, sillas, mesas))

# Mostrar resultados en tabla
print(f"\n{'Semana':<10} {'Sillas':>10} {'Mesas':>10}")
print("-" * 32)
for semana, s, m in historial:
    print(f"{semana:<10} {s:>10.2f} {m:>10.2f}")
```

```python
# Modelo logístico con cosecha constante
# P(n+1) = P(n) + r * P(n) * (1 - P(n)/K) - H

poblacion = float(input("Población inicial P_0: "))
n_años    = int(input("Años a simular: "))

# Parámetros del modelo
tasa_crecimiento = 0.4
capacidad_carga  = 1000
cosecha_anual    = 80

registros = [(0, poblacion)]

for año in range(1, n_años + 1):
    # Crecimiento logístico menos extracción
    nueva_pob = (poblacion
                 + tasa_crecimiento * poblacion * (1 - poblacion / capacidad_carga)
                 - cosecha_anual)
    poblacion = nueva_pob
    registros.append((año, poblacion))

print(f"\n{'Año':<8} {'Población':>12}")
print("-" * 22)
for año, p in registros:
    print(f"{año:<8} {p:>12.2f}")
```
